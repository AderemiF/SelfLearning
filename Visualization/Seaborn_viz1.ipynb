```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set(color_codes=True)
```

```python
print(sns.get_dataset_names())
```

```python
crash = sns.load_dataset('car_crashes')
```

```python
crash.head()
```

```python
crash.shape
```

```python
sns.displot(crash.not_distracted, kde=True, bins=25);
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash, kind='reg');
```

```python
sns.kdeplot(crash.speeding);
```

```python
sns.pairplot(crash);
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.shape
```

```python
tips.head()
```

```python
sns.pairplot(tips, hue='sex', palette='Blues');
```

```python
sns.rugplot(tips.tip);
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(8,4))
sns.set_context('paper', font_scale=1.5)
sns.jointplot(x='speeding', y='alcohol', data=crash, kind='reg');
```

```python
sns.set_style('dark')
plt.figure(figsize=(10,5))
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg');
```
